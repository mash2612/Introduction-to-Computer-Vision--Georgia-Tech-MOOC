## Contains the Code for Gaussian Noise

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.random.randn(10000) #Generating 100 columns of numbers from Gaussian Distribution 
#print(a)
```

```python
n, bins, patches = plt.hist(a, bins=[-3, -2, -1, 0, 1, 2, 3])
```

```python
print(bins,n, patches)
```

### Bins could be passed as a range instead of a list using linspace function from numpy

```python
n, bins, patches= plt.hist(a, bins=np.linspace(-3,3,7))
```

### To increase the resolution the number of bins can be increased

```python
n, bins, patches= plt.hist(a, bins=np.linspace(-3,3,21))
```

```python
n, bins, patches= plt.hist(a, bins=np.linspace(-3,3,100))
```

```python
gaussian_noise = np.random.randn(640,640)
```

```python
plt.imshow(gaussian_noise, cmap = 'gray')
```

```python
gaussian_noise_sig2 = 2*np.random.rand(640,640)
plt.imshow(gaussian_noise_sig2, cmap = 'gray')
plt.title('Gaussian Noise Sigma = 2')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.imshow(gaussian_noise, cmap='gray')
ax1.set_title('Normal Gaussian Noise')

ax2.imshow(gaussian_noise_sig2, cmap= 'gray')
ax2.set_title('Gaussian Noise with sigma = 2')
```
